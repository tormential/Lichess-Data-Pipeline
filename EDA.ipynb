```python
%%bash
# EDA

# 1. Dataset Overview
echo "=== Dataset Overview ==="
echo "Stockfish games:"
wc -l lichess_stockfish_sample.csv
echo "Non-stockfish games:"
wc -l lichess_nofish_sample.csv
```

```python
%%bash
# 2. CSV Structure
echo -e "\n=== Column Structure ==="
head -1 lichess_stockfish_sample.csv | tr ',' '\n' | nl
```

```python
%%bash
# 3. Game Results
echo -e "\n=== Game Results (Stockfish) ==="
cut -d',' -f8 lichess_stockfish_sample.csv | sort | uniq -c | sort -nr | head -3

echo -e "\n=== Game Results (Non-Stockfish) ==="
cut -d',' -f8 lichess_nofish_sample.csv | sort | uniq -c | sort -nr | head -3
```

```python
%%bash
# 4. Time Control Analysis
echo -e "\n=== Most Common Time Controls (Stockfish Analysis) ==="
cut -d',' -f11 lichess_stockfish_sample.csv | sort | uniq -c | sort -nr | head -10

echo -e "\n=== Most Common Time Controls (No Analysis) ==="
cut -d',' -f11 lichess_nofish_sample.csv | sort | uniq -c | sort -nr | head -10
```

```python
%%bash
# 5. Opening Analysis
echo "=== Most Frequent Openings (Stockfish Analysis) ==="
cut -d',' -f10 lichess_stockfish_sample.csv | tail -n +2 | sort | uniq -c | sort -nr | head -10
echo ""
echo "=== Most Frequent Openings (No Analysis) ==="
cut -d',' -f10 lichess_nofish_sample.csv | tail -n +2 | sort | uniq -c | sort -nr | head -10
```

```python
%%bash
# 6. Termination Analysis
echo "=== Game Termination ==="
echo "=== Without analysis: ==="
cut -d',' -f12 lichess_stockfish_sample.csv | tail -n +2 | sort | uniq -c | sort -nr | head -2

echo "=== Without analysis: ==="
cut -d',' -f12 lichess_nofish_sample.csv | tail -n +2 | sort | uniq -c | sort -nr | head -2
```

```python
%%bash
# 7. Sample Games
echo "=== Sample Games with Stockfish Analysis ==="
head -2 lichess_stockfish_sample.csv
echo -e "\n=== Sample Games without Stockfish Analysis ==="
head -2 lichess_nofish_sample.csv
```
